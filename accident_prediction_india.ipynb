## Handle missing values

### Subtask:
Implement a strategy for handling missing values in the DataFrame.


**Reasoning**:
Identify columns with missing values and determine the strategy for handling them. Impute missing values in 'Traffic Control Presence' and 'Driver License Status' with the mode.



```python
# Identify columns with missing values
missing_values = df.isnull().sum()
print("Missing values before handling:")
display(missing_values[missing_values > 0])

# Impute missing values in 'Traffic Control Presence' and 'Driver License Status' with the mode
for col in ['Traffic Control Presence', 'Driver License Status']:
    if df[col].isnull().sum() > 0:
        mode_value = df[col].mode()[0]
        df[col].fillna(mode_value, inplace=True)

# Verify that there are no remaining missing values in the DataFrame
print("\nMissing values after handling:")
display(df.isnull().sum()[df.isnull().sum() > 0])
```

## Data type conversion

### Subtask:
Convert columns to appropriate data types, especially 'Time of Day'.


**Reasoning**:
Convert the 'Time of Day' column to datetime objects and then verify the data types of all columns to ensure the conversion was successful.



```python
df['Time of Day'] = pd.to_datetime(df['Time of Day'], format='%H:%M:%S').dt.time

print("\nData types after converting 'Time of Day':")
df.info()
```

**Reasoning**:
The 'Time of Day' column was not converted to a time data type as intended, it's still an object type. The previous conversion attempt using `.dt.time` after `pd.to_datetime` keeps it as object type. I need to convert it to datetime and then extract the time component if needed for analysis, but for data type conversion, converting directly to datetime objects should be sufficient.



```python
df['Time of Day'] = pd.to_datetime(df['Time of Day'], format='%H:%M:%S')

print("\nData types after converting 'Time of Day' to datetime:")
df.info()
```

## Handle inconsistent data

### Subtask:
Address any inconsistencies in categorical columns.


**Reasoning**:
Identify and standardize inconsistent entries in categorical columns by converting them to lowercase and stripping whitespace.



```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"Unique values in '{col}' before cleaning:")
    print(df[col].unique())
    # Convert to lowercase and strip whitespace
    df[col] = df[col].str.lower().str.strip()
    print(f"Unique values in '{col}' after cleaning:")
    print(df[col].unique())
    print("-" * 30)
```

## Feature engineering

### Subtask:
Create new features like 'Hour of Day' or 'Season'.


**Reasoning**:
Extract the hour from 'Time of Day' and create a 'Season' column based on 'Month'.



```python
df['Hour of Day'] = df['Time of Day'].dt.hour

season_map = {
    'january': 'Winter', 'february': 'Winter', 'december': 'Winter',
    'march': 'Summer', 'april': 'Summer', 'may': 'Summer',
    'june': 'Monsoon', 'july': 'Monsoon', 'august': 'Monsoon',
    'september': 'Autumn', 'october': 'Autumn', 'november': 'Autumn'
}
df['Season'] = df['Month'].map(season_map)

display(df.head())
```

## Modelling

### Subtask:
Define Accident Severity as the target (classification problem).

Split data (train_test_split) into training and testing sets.

Choose models (start with Logistic Regression for interpretability, try others like Decision Tree, Random Forest).

Train and predict.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

# Define target and features
X = df.drop('Accident Severity', axis=1)
y = df['Accident Severity']

# Encode the target variable
le = LabelEncoder()
y = le.fit_transform(y)

# Select categorical and numerical columns for encoding
categorical_cols = X.select_dtypes(include='object').columns
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns

# Apply one-hot encoding to categorical features
X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Initialize and train Logistic Regression model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

# Predict on the test set
y_pred_lr = log_reg.predict(X_test)
```

### Subtask: Evaluate the model

Evaluate the performance of the trained Logistic Regression model using appropriate metrics like accuracy, precision, recall, F1-score, and a confusion matrix.
